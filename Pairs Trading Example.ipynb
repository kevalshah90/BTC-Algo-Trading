```python
import numpy as np
import pandas as pd
import statsmodels
import matplotlib.pyplot as plt
plt.figure(figsize=(2,2))
import statsmodels.tsa.stattools as ts
from statsmodels.tsa.stattools import coint
```

#### Let's generate two fake securities X and Y. The price or returns of Y are closely linked to X (For them to be cointegrated)

```python
 # Draw 100 samples of returns of X from a normal distribution with mean = 0 and std = 1 
X_returns = np.random.normal(0, 1, 100)
X = pd.Series(np.cumsum(X_returns), name='X') + 50
X.plot()
```

```python
# Add some random normally distributed noise to X to get the value of Y 
some_noise = np.random.normal(0, 1, 100)
Y = X + 5 + some_noise
Y.name = 'Y'
pd.concat([X, Y], axis=1).plot();
```

### Cointegration

Linear combination between two time series that will vary around the mean. 

```python
# Plot the spread 
(Y-X).plot()
plt.axhline((Y-X).mean(), color = 'red', linestyle='--')
plt.xlabel('Time')
plt.legend(['Price Spread', 'Mean']);
```

#### Testing for cointegration 

```python
score, pvalue, _ = coint(Y,X)
print('Score:', str(score),'\n','P-Value:', str(pvalue))
```
